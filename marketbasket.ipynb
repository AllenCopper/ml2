```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
from mlxtend.preprocessing import TransactionEncoder
```

```python
data=[]
with open("./Assignment-5.csv") as file:
    reader = csv.reader(file,delimiter=',')
    for row in reader:
     data+=[row]
```

```python
data[:5]
```

```python
te = TransactionEncoder()
te
```

```python
xe = te.fit_transform(data)
xe
```

```python
df = pd.DataFrame(xe,columns=te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
freq_item = apriori(df,min_support=0.02,use_colnames=True)
freq_item
```

```python
asso = association_rules(freq_item,metric="confidence",min_threshold=0.02)
```
